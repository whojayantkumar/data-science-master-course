```python
import bs4
import requests
url = "https://www.passiton.com/inspirational-quotes?page=2"
```

```python
response = requests.get(url)
soup = bs4.BeautifulSoup(response.content)
print(soup)
```

```python
div_element = soup.findAll('div')
div_element[0]
```

```python
div_element[0].div
```

```python
for div in div_element:
    anchor = div.findAll('img').attr
    print(anchor)
```
