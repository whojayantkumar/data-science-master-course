```python
from bs4 import BeautifulSoup as soup
windows_version_history_soup = soup(windows_version_history_html,'html.parser')
print(windows_version_history_soup)
```

```python
windows_version_history_soup.h1
```

```python
tables = windows_version_history_soup.findAll('table',{'class' : 'wikitable sortable'})
windows_version_history_table = tables[0]
print(windows_version_history_table)
```

```python
#column titles
headers = windows_version_history_soup.findAll('th',{})
headers = headers[0:7]
```

```python
column_titles = [ct.text[:-1] for ct in headers]
print(column_titles)
```

```python
# rows data
rows_data = windows_version_history_table.findAll('tr')[1:]
print(rows_data)
```

```python
# first row
first_row = rows_data[0].findAll('td')
print(first_row)
for td in first_row:
    print(td.text[-1])
```

```python
# create table rows
table_rows = []
for row in rows_data:
    current_row = []
    row_data = row.findAll('td')
    for idx,data in enumerate(row_data):
        current_row.append(data.text[:-1])
    table_rows.append(current_row)
print(table_rows)
```

```python
# write data in csv file
filename = "windows_version_history.csv"
with open(filename,'w',encoding='utf-8') as f:
    # write column titles in csv file
    header_string = ','.join(column_titles) + '\n'
    f.write(header_string)
    
    # write rows data in file
    for row in table_rows:
        row_string = ''
        for w in row:
            w = w.replace(',','')
            row_string += w + ','
        row_string +=  '\n'
        f.write(row_string)
```
